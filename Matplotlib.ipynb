```python
#Diagrama de areas
import matplotlib.pyplot as plt
#crea cuadro de figura
fig, ax = plt.subplots()

ax.fill_between([1,2,5,0],[2,0,4,1])
#guardar el grafico
plt.savefig('Diagrama_prueba_area.png')

#mostrar el grafico
plt.show()
```

```python
#Diagrama de histograma
import matplotlib.pyplot as plt
import numpy as np

#crea cuadro de figura
fig, ax = plt.subplots()

x=np.random.normal(5,1.5,size=1000)
ax.hist(x,np.arange(0,11))
#guardar el grafico
plt.savefig('Diagrama_prueba_histograma.png')

#mostrar el grafico
plt.show()
```

```python
#Diagrama de diagrama de sectores
import matplotlib.pyplot as plt
import numpy as np

#crea cuadro de figura
fig, ax = plt.subplots()

ax.pie([5,4,3,2,1])
#guardar el grafico
plt.savefig('Diagrama_prueba_diagrama de sectores.png')

#mostrar el grafico
plt.show()
```

```python

#Diagrama de diagrama de sectores
import matplotlib.pyplot as plt


peso = []
nombre = []

for j in range (0,5):
  nom=input(f"nombre de estudiante {j+1}  ")
  nombre.append(nom)

  pesos_estudiantes=float(input(f"peso: {j+1}   "))
  peso.append(pesos_estudiantes)
  promedio=sum(peso)/len(peso)


fig, ax = plt.subplots()
# autopct se usa para mostrar porcentajes
ax.pie(peso, labels=nombre, autopct='%1.1f%%', startangle=90)




print(f"proemdio de peso en kg {promedio}")

#guardar el grafico
plt.savefig('Diagrama_prueba_peso.png')

#mostrar el grafico
plt.show()

```

```python
#diagrama de contorno
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z)
plt.show()
```
